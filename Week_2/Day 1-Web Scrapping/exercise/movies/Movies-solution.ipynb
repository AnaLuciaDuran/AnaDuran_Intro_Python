```python
import pandas as pd
import numpy as np

data = pd.read_csv("movies.csv", header=None, names =["Movie","Year","Box_Office"])
```

```python
df = pd.DataFrame (data)
df.head(10)
```

```python
def most_succesfull (number_of_movies, max_year, df):

    filter_df = df[df["Year"] <= max_year]
    print(df)
    filter_df = df.sort_values(by=["Box_Office"], ascending = False)
      
    return filter_df.iloc[ :number_of_movies]
```
