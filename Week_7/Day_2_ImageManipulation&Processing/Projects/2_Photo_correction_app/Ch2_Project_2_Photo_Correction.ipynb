1. Create a function that given a photo of your notes with bad illumination it changes the image so it can be read

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def showImg(img, title = ''):
    plt.figure(figsize = (10,5));
    plt.title(title)
    plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))

def showGrayScaleImg(img, title = ''):
    plt.figure(figsize = (10,5));
    plt.title(title)
    plt.imshow(img, cmap='gray')
```

```python
note=cv2.imread('note.jpg')
showImg(note)
```

```python
note2=cv2.imread('note.jpg',0)
showImg(note2)
th2 = cv2.adaptiveThreshold(note2,255,cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY,31,3)
showImg(th2)
```

```python
note_copy = note.copy()
for i in range(2):
    note_copy = cv2.pyrUp(note_copy)
    showImg(note_copy)
```

```python
pixel_values_matrix = np.ones(note_copy.shape, dtype= "uint8") * 50
brightened_image = cv2.add(note_copy,pixel_values_matrix)
showImg(brightened_image)
```

```python
gray_note = cv2.cvtColor(brightened_image, cv2.COLOR_BGR2GRAY)
showImg(gray_note)
```

```python
negative_image = cv2.bitwise_not(gray_note)
ret, thresh = cv2.threshold(gray_note,0,255,cv2.THRESH_BINARY | cv2.THRESH_OTSU)
showGrayScaleImg(thresh)
```

```python

th2 = cv2.adaptiveThreshold(note2,255,cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY,35,3) 

th3 = cv2.adaptiveThreshold(note2,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,35,3)

titles = ['ADAPTIVE_THRESH_MEAN_C','ADAPTIVE_THRESH_GAUSSIAN_C']
images = [th2,th3]
for i in range(2):
    showImg(images[i],titles[i])
```

```python
note=cv2.imread('note.jpg')
note2 = cv2.cvtColor(note, cv2.COLOR_BGR2GRAY)
```

```python
threshold = 200
max_value= 255

ret5, thresh5 = cv2.threshold(note2, threshold, max_value,
                              cv2.THRESH_TOZERO_INV)

showImg(thresh5)
```

```python
th3 = cv2.adaptiveThreshold(thresh5,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,35,7)
showImg(th3)
```
