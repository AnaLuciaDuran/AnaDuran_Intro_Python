I have a linear equation: Y = mX + b




```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
sns.set()
```

```python
df=pd.read_csv('reg_data.csv',sep=',', 
                  names=['x', "y"])
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
points = df.values
```

```python
x = points[:,0]
y = points[:,1]
```

```python
n = len (x)
sumx = np.sum(x)
sumy = np.sum(y)
sumx2 = np.sum(x*x)
sumy2 = np.sum(y*y)
sumxy = np.sum(x*y)
mean_x = np.mean (x) 
mean_y = np.mean (y)
sumx, sumy, sumx2, sumy2, sumxy, mean_x, mean_y
```

I have a linear equation: Y = mX + b




```python
m = (sumx*sumy-n*sumxy)/(sumx**2-n*sumx2)
b = mean_y-m*mean_x
m , b
```

```python
sigmax = np.sqrt(sumx2/n-mean_x**2)
sigmay = np.sqrt(sumy2/n-mean_y**2)
sigmaxy = sumxy/n-mean_x*mean_y
R2 = (sigmaxy/(sigmax*sigmay))**2
R2
```

```python
plt.plot(x,y, 'o', label='data')
plt.plot (x, m*x+b, label='LR')
plt.xlabel ('x')
plt.ylabel ('y')
plt.title('My first lineal regression in python')
plt.grid()
plt.legend()
plt.show()
```
