# Define the functions required

## Function 1:

It is a function to determine or estimate the coefficients where x and y values are passed into this function.

```python
def estimate_coefficients (x,y):
    mean_x, mean_y = np.mean(x), np.mean(y)
    temp = x - mean_x
    b1 = np.sum(temp * (y-mean_y)) / np.sum(temp**2)
    b0 = mean_y - b1 * mean_x
    return (b0,b1)
```

```python
def plot_regression_line (x,y,b):
    plt.scatter (x,y, color = 'red', marker = 'o', s = 30)
    y_pred = b[0]+b[1]*x
    plt.plot (x, y_pred, color = 'black')
    plt.xlabel('Mi x')
    plt.ylabel('Mi y')
    plt.show()
```
